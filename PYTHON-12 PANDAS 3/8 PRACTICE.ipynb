Для решения задач нам понадобится выделить из признака title год выпуска фильма. Для этого напишем функцию get_year_release(arg).

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        # если число вхождений больше 0,
	    # очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

[**Модуль re**](https://tproger.ru/translations/regular-expression-python/ "Подробнее о регулярных выражениях")

предназначен для поиска шаблонов в тексте и встроен в язык, поэтому не нуждается в установке.

Из модуля re нам понадобится только функция **findall()**, которая позволяет найти в строке все слова, удовлетворяющие шаблону. Мы находим в строке с названием фильма шаблон "(DDDD)" — четыре цифры, обёрнутых в скобки, что соответствует году выпуска фильма. Если такого шаблона не было найдено (год выпуска не указан), функция возвращает None (в таблице это будет помечено как пропуск).

```python
# Создайте в таблице новый признак year_release, который соответствует году выпуска фильма.
# У скольких фильмов не указан год их выпуска?
ratings_movies['year_release'] = ratings_movies['title'].agg(get_year_release)
display(ratings_movies['year_release'].notna().value_counts())
```

```python
# Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей?
# В качестве ответа запишите название этого фильма без указания года его выпуска.
mean_rating_1999 = ratings_movies[ratings_movies['year_release'] == 1999].groupby('title')['rating'].mean().sort_values()
display(mean_rating_1999)
```

```python
# Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)?
# Запишите сочетание так же, как оно указано в таблице (через разделитель |, без пробелов)
rating_genres_2010 = ratings_movies[ratings_movies['year_release'] == 2010].groupby('genres')['rating'].mean().sort_values()
display(rating_genres_2010)
```

```python
# Какой пользователь (userId) посмотрел наибольшее количество различных (уникальных) жанров
# (genres) фильмов? В качестве ответа запишите идентификатор этого пользователя
unique_genres_user = ratings_movies.groupby('userId')['genres'].nunique().sort_values(ascending=False)
display(unique_genres_user)
```

```python
# Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка фильмам наибольшая.
# В качестве ответа укажите идентификатор этого пользователя.
user_leastratings_mostmean = ratings_movies.groupby('userId')['rating'].agg(['count','mean']).sort_values(by='mean', ascending=False)
display(user_leastratings_mostmean)
```

```python
# Найдите сочетание жанров (genres) за 2018 год, которое имеет наибольший средний рейтинг (среднее по столбцу rating),
# и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10.
# Запишите сочетание так же, как оно указано в таблице (через разделитель |, без пробелов)
genres_2018_mostrating10 = ratings_movies[ratings_movies['year_release'] == 2018].groupby('genres')['rating'].agg(['mean', 'count']).sort_values(by=['mean','count'],ascending=[False,False])
display(genres_2018_mostrating10[genres_2018_mostrating10['count'] > 10])
```

```python
# Добавьте в таблицу новый признак year_rating — год выставления оценки.
# Создайте сводную таблицу, которая иллюстрирует зависимость среднего рейтинга фильма
# от года выставления оценки и жанра. Выберите верные варианты ответа, исходя из построенной таблицы:
ratings_movies['year_rating'] = pd.to_datetime(ratings_movies['date']).dt.year
pivot = ratings_movies.pivot_table(
    values='rating', 
    index='year_rating',
    columns='genres',
    aggfunc='mean',
)
display(pivot)

# За весь период (с 1996 по 2018 год) сочетание жанров Action|Adventure ни разу не получало среднюю оценку ниже 3.
pivot['Action|Adventure'].min()

# Наилучшую оценку жанр Action|Adventure|Animation|Children|Comedy|IMAX получил в 2010 году.
pivot['Action|Adventure|Animation|Children|Comedy|IMAX'].nlargest()

# Среди сочетаний жанров, получивших наивысшую среднюю оценку в 2018 году, есть сочетание Animation|Children|Mystery.
pivot.loc[2018,'Animation|Children|Mystery']

# Для жанра Comedy прослеживается тенденция падения рейтинга с каждым годом (с 1996 по 2018)
pivot['Comedy']
```

Дано два файла:

* orders.csv, содержащий данные о заказах;
* products.csv, содержащий данные о товарах.

Объедините заданные таблицы в таблицу orders_products, чтобы в результирующей таблице оказалась информация обо всех заказах, но не оказалось информации о продуктах, на которые заказов ещё не поступало.

```python
orders = pd.read_csv('data/orders/orders.csv', sep=';')
#display(orders)
products = pd.read_csv('data/orders/products.csv', sep=';')
#display(products)
orders_products = orders.merge(
    products,
    left_on='ID товара',
    right_on='Product_ID',
    how='left'
)
orders_products = orders_products.drop('Product_ID',1)
display(orders_products.tail())
```

```python
# Какой идентификатор (Order ID) имеет заказ, для которого не оказалось информации о товаре
orders_products[orders_products['Name'].isna()]
```

```python
# На какой товар была произведена отмена?
# В качестве ответа запишите название этого товара (Name)
orders_products[orders_products['Отменен'] == 'Да']['Name']
```

```python
# Какой покупатель принёс наибольшую суммарную прибыль интернет-магазину за указанный период?
# В ответ запишите идентификатор этого покупателя (ID Покупателя).
# Прибыль состоит только из оплаченных заказов и рассчитывается как количество купленного товара, умноженное на его цену.
orders_products['profit'] = orders_products['Price']*orders_products['Количество']
grouped = orders_products[orders_products['Оплачен'] == 'Да'].groupby('ID Покупателя')['profit'].agg('max').sort_values(ascending=False)
display(grouped)
```
