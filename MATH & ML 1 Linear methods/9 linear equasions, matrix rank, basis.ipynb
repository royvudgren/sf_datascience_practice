## **РАНГ МАТРИЦЫ. БАЗИС**

Давайте посмотрим на преобразованные ступенчатые матрицы, которые у нас получились в двух прошлых примерах:

В первом случае у нас есть две нулевые строки. Таким образом, во всей матрице независимых столбцов будет два — это те столбцы, которые стоят в ступеньках. Отсюда следует, что все три столбца (векторы , v1, v2 и v3) являются линейно зависимыми между собой.

Во втором случае все три столбца составляют ступени, поэтому все они будут линейно независимы.
***
* **Ранг матрицы** — это количество линейно независимых столбцов. Обозначение: **rk** или **rank**.

* **Ранг системы векторов** — это размерность этой системы.
***

Получается, чтобы провести исследование системы векторов v1,v2, ... vn, каждый вектор размерности dim(v) = n на линейную зависимость при помощи ранга, нам достаточно записать их в виде матрицы

![](data/134.PNG)

и найти её ранг rk(A). Он будет равен максимальному количеству линейно независимых векторов в системе.

* Если ранг равен количеству векторов в системе (rk(A) = n), то система векторов является **линейно независимой**.
* Если ранг меньше, чем количество векторов в системе (rk(A) < n), можно смело говорить, что система векторов является линейно зависимой. В линейно независимой системе нельзя выразить один из векторов через остальные, в зависимой — можно, однако не любой. 

Знание ранга даёт нам только количество линейно независимых векторов, но не говорит, какие именно векторы выражаются или не выражаются друг через друга. Для ответа на этот вопрос необходимо найти **базис системы**.
***

* **Базис** — это такая часть системы векторов v1,v2, ... vk, через которую можно выразить все остальные векторы k<=n. 
***
В задаче регрессии это может быть такой набор факторов (столбцов исходной таблицы с данными), через который выражаются все остальные факторы. Например, через факторы жилой и общей площади квартиры можно линейно выразить фактор нежилой площади квартиры. Аналогично и жилую площадь можно выразить через общую и нежилую. 

Более формальное определение базиса будет выглядеть так: 

                Базис — это максимальная линейно независимая подсистема векторов v1,v2, ... vn.

Это логично, ведь сначала мы берём один вектор, далее добавляем ещё один и делаем так до тех пор, пока система не будет оставаться линейно зависимой, какой бы вектор мы ни добавляли.

Количество векторов в базисе совпадает с рангом системы. При этом самих базисов может быть несколько, мы выбираем их произвольно.

Давайте рассмотрим пример того, как вычислить ранг в Python. Для этого в библиотеке numpy существует специальная функция **linalg.matrix_rank()**:

![](data/135.PNG)

**Решение на Python**

```python
import numpy as np
v1 = np.array([1,2,3,1])
v2 = np.array([4,5,6,1])
v3 = np.array([7,8,9,1])
A = np.array([v1, v2, v3]).T
print(np.linalg.matrix_rank(A))
```

Мы видим, что ранг системы равен 2, а векторов три — значит система является линейно зависимой. В качестве базиса можем взять любые два вектора. Очевидно, каждая пара здесь является линейно независимой, т. к. никакие два вектора не пропорциональны.

Скажем, например, что базисом системы векторов v1, v2, v3 будут векторы v1 и v2. 

Решим второй пример, в котором изменена координата вектора v3:

![](data/136.PNG)

```python
import numpy as np
v1 = np.array([1,2,3,1])
v2 = np.array([4,5,6,1])
v3 = np.array([7,8,10,1])
A = np.array([v1, v2, v3]).T
print(np.linalg.matrix_rank(A))
```

В данном примере ранг равен 3 — значит векторы линейно независимы и в базис войдут все три вектора.

```python
v1 = np.array([1,0,3,5])
v2 = np.array([0,4,5,5])
v3 = np.array([0,0,0,4])
v4 = np.array([0,0,0,0])
A = np.array([v1, v2, v3, v4])
A
print(np.linalg.matrix_rank(A))
```
